```python
# after skipping the first row which was attribute name, next row element now becomes the attributes
# skipping two rows
df = pd.read_csv("G:\\datasets\\student_results.csv", skiprows = 2)
df
```

```python
# two rows have been skipped, first row is attribute name or header and second row was the element 

# now let me skip the specific rows, and for this skiprows will take the parameter as list which element will
# be the index of the rows
df = pd.read_csv("G:\\datasets\\student_results.csv", skiprows = [0]) # 0 index row was the header names, which will get skipped
df
```

```python
df = pd.read_csv("G:\\datasets\\student_results.csv")
df
```

```python
# now skipping the row
# 3 1004 11 0 10 1 5 45
# its index will be 4, start index counting from the header names
df = pd.read_csv("G:\\datasets\\student_results.csv", skiprows = [4])
df
```

```python
# for skipping the multiple rows give the index number of that rows in skiprows
df = pd.read_csv("G:\\datasets\\student_results.csv", skiprows = [2,3,5])
df
```

```python
# here i am repeadly getting the index number from the python
# but now i want to make my any one column as index 
# for that we will use index_col
df = pd.read_csv("G:\\datasets\\student_results.csv", index_col = 'Student ID')
df
```
