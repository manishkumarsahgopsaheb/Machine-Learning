```python
# now lets check is there any null value is in the dataset
# for this we have isnull() and notnull()
# lets implement this
df = pd.read_csv("G:\\datasets\\student_results.csv")
df
```

```python
# now checking for the null value
df.isnull()
```

```python
# in the above dataset 
# False represent -- it is not null
# True  represent -- it is null

# now lets check which column contain how many null values
df.isnull().sum()
```

```python
# see in the above output class contain 2 null value and percentage column also contain 2 null values

# now lets check the total null values. i mean overall
df.isnull().sum().sum()
```

```python
# so overall we have 4 null values
```

```python
# now lets use notnull()
df.notnull()
```

```python
# here True represent --- not null values
# and  False represent --- null values
```

```python
# sameway we can also calcu;late total not null value
df.notnull().sum()
```

```python
# for overall notnull
df.notnull().sum().sum()
```

```python
# so in total 76 values are there in the dataset which are not nul
```
